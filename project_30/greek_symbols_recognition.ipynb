#### importing images to python environment

```python
from glob import glob 
import os

images_ar = []

class_names = os.listdir('data_source\symbols')
print(class_names)

for folder_name in class_names:
    path = 'data_source/symbols/{}/*.png'.format(folder_name)
    for file_name in glob(path):
        img = Image.open(file_name)
        
        img_resized = img.resize((30, 30))
        img_resized = np.array(img_resized)
        images_ar.append(img_resized)
        
        img_hstr = img.resize((40, 30))
        img_hstr_resized = img_hstr.resize((30, 30))
        img_hstr_resized = np.array(img_hstr_resized)
        images_ar.append(img_hstr_resized)
        
        img_vstr = img.resize((30, 40))
        img_vstr_resized = img_vstr.resize((30, 30))
        img_vstr_resized = np.array(img_vstr_resized)
        images_ar.append(img_vstr_resized)
        
        img_clrt = img.rotate(15)
        img_clrt_resized = img_clrt.resize((30, 30))
        img_clrt_resized = np.array(img_clrt_resized)
        images_ar.append(img_clrt_resized)
        
        img_aclrt = img.rotate(-15)
        img_aclrt_resized = img_aclrt.resize((30, 30))
        img_aclrt_resized = np.array(img_aclrt_resized)
        images_ar.append(img_aclrt_resized)
        
        

images_ar = np.array(images_ar)        
print(images_ar.shape)

label_ar = []
for i in range(0, 9):
    for j in range(0, 75):
        label_ar.append(i)
        
label_ar = np.array(label_ar)
print(label_ar)
print(label_ar.shape)

# plt.figure(figsize=(3, 3))
# plt.imshow(img)
# plt.title('vizualization of image', fontsize=9)
# plt.xticks(fontsize=8); plt.yticks(fontsize=8)
# plt.show()
```

#### convert images and labels to pandas data frame

```python
# convert feature and labels to pandas data frame
import pandas as pd
pd.set_option('display.max_columns', None)

list_of_tuples = list(zip(images_ar, label_ar))
df = pd.DataFrame(list_of_tuples, columns=['features', 'labels'])

display(df)
```

#### shuffling the feature-label rows

```python
from sklearn.utils import shuffle
df_sfl = shuffle(df); df_sfl = df_sfl.reset_index(drop=True)
display(df_sfl)
```

#### vizualizing the features and labels

```python
plt.figure(figsize=(10, 6))
for i in range(1, 25):
    plt.subplot(4, 6, i)
    plt.imshow(df_sfl['features'][i-1])
    plt.xlabel(class_names[df_sfl['labels'][i-1]], fontsize=9)
    plt.xticks(fontsize=8), plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

#### train test split and converting features to tensors

```python
X = df_sfl[['features']]; y = df_sfl[['labels']]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=11)
print(len(X_train), len(y_train), len(X_test), len(y_test))

# --create tensors to be usable by a model
import tensorflow as tf

X_train = np.array(X_train).tolist()
X_train = tf.Variable(X_train, tf.float64)
X_train = tf.reshape(X_train, [506, 30, 30, 3])
print(X_train.shape)

X_test = np.array(X_test).tolist()
X_test = tf.Variable(X_test, tf.float64)
X_test = tf.reshape(X_test, [169, 30, 30, 3])
print(X_test.shape)
```

### buildng the neural network model

```python
# --building the neural network
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Flatten(input_shape=(30, 30, 3)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(9)
])

adam = keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=False)

model.compile(
    optimizer=adam,
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

model.summary()
```

#### testing the model

```python
from tensorflow.keras.models import load_model
model = load_model('greek_symbol_recognition.h5')

predictions_prob = model.predict(X_test)

predictions = []
predictions_class = []
for x in predictions_prob:
    ar = np.array(x)
    max_val = np.amax(ar)
    index = np.where(ar == max_val)[0][0]
    predictions.append(index)
    predictions_class.append(class_names[index])

print(predictions_class[10:20])

actual_class = []
for x in y_test['labels']:
    actual_class.append(class_names[x])
    
print()    
print(actual_class[10:20])    

# getting the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print()
print('accuracy =', accuracy)
```

### using the model

```python
import numpy as np
from glob import glob
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from PIL import Image
import os

# --configurations
%config InlineBackend.figure_formats = ['svg'] # --for clear matplotlib plots 

import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.models import load_model
model = load_model('greek_symbol_recognition.h5')

class_names = os.listdir('data_source\symbols')

images_ar = []
for file_name in glob('data_source/test/*.png'):
        img = Image.open(file_name)
        img = img.resize((30, 30))
        img = np.array(img).tolist()
        images_ar.append(img)
        
images_ar = np.array(images_ar)
images_ar_t = tf.constant(images_ar, tf.float64)
print(images_ar_t.shape)
        
predictions_prob = model.predict(images_ar_t)

predictions_class = []
for x in predictions_prob:
    ar = np.array(x)
    max_val = np.amax(ar)
    index = np.where(ar == max_val)[0][0]
    predictions_class.append(class_names[index])

print(predictions_class)

plt.figure(figsize=(10, 6))
for i in range(0, len(images_ar_t)):
    plt.subplot(4, 6, i+1)
    plt.imshow(images_ar[i])
    plt.xlabel(predictions_class[i], fontsize=9)
    plt.xticks(fontsize=8), plt.yticks(fontsize=8)
    
plt.tight_layout()
plt.show()
```
